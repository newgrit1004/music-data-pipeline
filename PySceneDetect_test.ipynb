```python
from func.video_preprocessing import crop_video
from typing import Tuple
def crop_song_title_video(song_title:str, coordinates:Tuple[int]):
    lyric_video_save_folder = './lyric_video'
    video_file_name = os.listdir(os.path.join(lyric_video_save_folder, song_title))[0]
    video_file_path = os.path.join(lyric_video_save_folder, song_title, video_file_name)
    cropped_video_path = '/'.join(['cropped_video', song_title, f'{song_title}.mp4'])
    os.makedirs('/'.join(['cropped_video', song_title]), exist_ok=True)
    crop_video(video_file_path, coordinates, cropped_video_path)
```

```python
import subprocess
def convert_grayscale_video(song_title:str):
    """cropped_video에 저장된 비디오들을 gray_scale로 변환하는 함수이며 변환하는 이유는 OCR 성능을 올리기 위해서임."""
    cropped_video_save_folder = './cropped_video'
    grayscale_folder = './gray_cropped_video'
    video_file_name = os.listdir(os.path.join(cropped_video_save_folder, song_title))[0]
    video_file_path = os.path.join(cropped_video_save_folder, song_title, video_file_name)

    song_video_file_path = video_file_path.split('/')[-1]
    gray_video_folder = '/'.join((grayscale_folder, song_title))

    gray_video_file_path = '/'.join((gray_video_folder, song_video_file_path))
    os.makedirs(gray_video_folder, exist_ok=True)
    cmd = ['ffmpeg', '-i', '%s'%str(video_file_path), '-vf', 'format=gray', '%s'%str(gray_video_file_path)]
    subprocess.run(cmd, shell=False)
```
